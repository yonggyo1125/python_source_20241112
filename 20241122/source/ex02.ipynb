```python
faker = Faker("ko-KR")
```

```python
faker.name()
```

```python
members = [(faker.email(), faker.name(), faker.address()) 
            for _ in range(0, 1000)] # i 변수가 사용 되지 않으면 _로 자리만 채워준다!

for member in members:
    print(member)
```
